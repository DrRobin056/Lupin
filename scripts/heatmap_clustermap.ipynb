****Generating Simple Heatmaps or Clustermaps from 2D data****

Simple code to begin generating heatmaps or cluster maps from the SNP data in a cohort of patients. 

**Heatmaps** 

This is the simplest for of heat map based on a 2D matrix. Note that the order of columns will be preserved. Therefore this could be a good way of mainating meaningful clusters which have been ascertained by the clustering algorithms. The data would need to be ordered by cluster label - CD-ptcluster.tsv. This could help to distinguish the cluster defining SNPs. 

This would be the preferred method.

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np

#Create dataframe for either method

df_ptdata = pd.read_csv('/Users/rb1425/Library/CloudStorage/OneDrive-ImperialCollegeLondon/PHD/LUPIN/LUPIN_code/output files/CD_affected_atg_patients.tsv', sep='\t')
df_ptdata.set_index('Affected protein', inplace=True)
df_ptdata
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df_ptdata, cmap="viridis", annot=True)
plt.title("Patients Clustered by SNP") 
plt.show()
```

**Clustermaps** 

This will hierachically cluster data in the form of a dendrogram. The algorithm can in some ways be customised - e.g. the choice of distance metric. Note we can colour the data but cluster data - try to assing this to the data later.  

```python
#simple clustermap 
sns.set_style("whitegrid")
plt = sns.clustermap(df_ptdata, yticklabels=True, figsize=(40,30), metric='jaccard')
```
